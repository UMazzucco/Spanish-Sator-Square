```python
five_letter_words = list(open('diccionario-espanol-txt-master/length/05.txt'))
print(five_letter_words,'\n',len(five_letter_words))
```

```python
corrected_five_letter_words = []
for i in five_letter_words:
    a = i.replace('\n','').replace('á','a').replace('é','e').replace('í','i').replace('ó','o').replace('ú','u')
    if a not in corrected_five_letter_words: 
        corrected_five_letter_words.append(a)
print(corrected_five_letter_words,'\n',len(corrected_five_letter_words))
```

```python
reversibles = []
for a in corrected_five_letter_words:
    b = a[::-1]
    if b in corrected_five_letter_words:
        reversibles.append(a)
print(reversibles,'\n',len(reversibles))
```

```python
palindromes = []
for i in corrected_five_letter_words:
    if (i[0]==i[4] and i[1]==i[3]):
        palindromes.append(i)
print(palindromes,'\n',len(palindromes))
```

```python
nonpal = []
for i in reversibles:
    if not (i[0]==i[4] and i[1]==i[3]):
        nonpal.append(i)
print(nonpal,'\n',len(nonpal))
```

```python
sator = []
for a in palindromes:
    for b in nonpal:
        if b[2] == a[0]:
            for c in nonpal:
                if (c[2]==a[1] and c[0]==b[1] and c[4]==b[3]):
                    e = [b,c,a,c[::-1],b[::-1]]
                    sator.append(e)
print(len(sator))
for i in sator:
    print(i[2],':',i)
```
